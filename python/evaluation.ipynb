```python

#filter dataframes
surveys['createdAt'] = pd.to_datetime(surveys['createdAt'])
surveys = surveys[(surveys['createdAt'] > '2021-07-12')]

suffix = surveys.groupby(surveys.hash).cumcount().astype(str).str.replace('0', '').values #add iterator to hash to differentiate multiple rows from same author (in order to assign multiple surveys)
surveys.hash = surveys.hash + suffix


memories['createdAt'] = pd.to_datetime(memories['createdAt']).dt.date
memories['date'] = pd.to_datetime(memories['date']).dt.date
suffix = memories.groupby(memories.hash).cumcount().astype(str).str.replace('0', '').values
memories.hash = memories.hash + suffix

interactions['createdAt'] = pd.to_datetime(interactions['createdAt'])
interactions = interactions[(interactions['createdAt'] > '2021-07-20')] #ignore interactions that happened before the public release
#interactions = interactions[(interactions['event'] != 'sessionEnd')] # roptionally emove in order to really look for last active interaction (intead of e.g. closing a tab later)

#create list of hashes of sessions that start too smal and filter all following events
startTooSmallHashes = interactions[(interactions['event'].str.startswith('notLargeEnough'))]['hash'].values.tolist()
numberOfSessions = interactions.loc[interactions['event'] == 'sessionStart']["id"].count()
print(f'filtered {len(startTooSmallHashes)} sessions with too small screen from {numberOfSessions} sessions total ({round(len(startTooSmallHashes)/numberOfSessions,2)*100} %)')
interactions = interactions[~interactions['hash'].isin(startTooSmallHashes)]

#filter sessions that did not consent
numberOfSessions = interactions.loc[interactions['event'] == 'sessionStart']["id"].count()
didNotConsent = []
for hash, group in interactions.groupby('hash'):
    if "consent" not in group["event"].to_list():
        didNotConsent.append(hash)

print(f'filtered {len(didNotConsent)} sessions with no consent from {numberOfSessions} remaining sessions total ({round(len(didNotConsent)/numberOfSessions,2)*100} %)')
interactions = interactions[~interactions['hash'].isin(didNotConsent)]

numberOfSessions = interactions.loc[interactions['event'] == 'sessionStart']["id"].count()
print(f'{numberOfSessions} unique sessions remain')

yesterday = date.today() - timedelta(days=1)
sessionsSinceYesterday = interactions[(interactions['createdAt'] > str(yesterday)) & (interactions['event'] == "sessionStart")]
print(f'{sessionsSinceYesterday.shape[0]} new sessions since yesterday')

```
